```python
from IPython.display import clear_output
import random

def display_board(board):
    clear_output()
    print('   |'*2)
    print(' '+board[1]+' | '+board[2]+' | '+board[3])
    print('   |'*2)
    print('-'*11)
    print('   |'*2)
    print(' '+board[4]+' | '+board[5]+' | '+board[6])
    print('   |'*2)
    print('-'*11)
    print('   |'*2)
    print(' '+board[7]+' | '+board[8]+' | '+board[9])
    print('   |'*2)

def player_input():
    print('Welcome to Tic Tac Toe!\n')
    symb=''
    while symb!='X' and symb!='O':
        symb=input('Player 1: Do you want to be X or O? ')
        if symb!='X' and symb!='O':
            print('Invalid symbol!')   
    return symb

def place_marker(board, marker, position):
    board[position]=marker
    
def win_check(board, mark):
    for i in range(2,9,3):
        if board[i-1]==board[i]==board[i+1]==mark:
            return True
    for i in range(4,7):
        if board[i-3]==board[i]==board[i+3]==mark:
            return True
    if board[1]==board[5]==board[9]==mark:
        return True
    if board[3]==board[5]==board[7]==mark:
        return True
    return False

def choose_first():
    player=random.randint(1,2)
    print(f'Player {player} will go first.\n')
    return player

def space_check(board, position):
    return board[position]==' '

def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True

def player_choice(board):
    pos=0
    while (not space_check(board, pos)):
        pos=int(input('Choose your next position: (1-9) '))
        if (not space_check(board, pos)):
            print('Position is occupied!')
    return pos

def replay():
    ans=''
    while ans!='Yes' and ans!='No':
        ans=input('Do you want to play againt? Enter Yes or No: ')
        if ans!='Yes' and ans!='No':
            print('Invalid response! (WARNING: Case sensitive)')
    return ans=='Yes'
```

```python
while True:
    one=player_input()
    if one=='X':
        two='O'
    else:
        two='X'
    turn=choose_first()
    input('Enter any character to start')
    board=[' ']*10
    board[0]='#'
    while not full_board_check(board):
        display_board(board)
        if win_check(board, 'O') or win_check(board, 'X'):
            print('Congratulations! You have won the game!')
            break
        #Player 1 Turn
        if turn==1:
            place_marker(board, one, player_choice(board))
            turn=2
        # Player2's turn.
        else:
            place_marker(board, two, player_choice(board))
            turn=1
    if not (win_check(board, 'X') or win_check(board, 'O')):
        print('The game ended in a draw.')
    if not replay():
        break
```
